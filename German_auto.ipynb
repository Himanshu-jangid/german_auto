```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
cars = {"Mileage":[10, 12, 13, 14, 5, 6],
          "Price": ["1000000", "3000000", "2500000", "3500000", "4500000", "5000000"],
          "sales":[1000, 200, 150, 400, 50, 60],
          "HorsePower":["120", "150", "170", "200", "550", "650"]
          }
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.DataFrame(cars)
x = df["Price"]
y = df["Mileage"]
plt.bar(x,y, label="Price", color="Red")
plt.xlabel("Price")
plt.ylabel("Mileage")
plt.title("German_Auto")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

Mileage= [22,45,30,59,58,56,57,45,43,43,43,50,40,34,33,25,19]
interval = [0,10,20,30,40,50,60]
plt.hist(Mileage, interval, histtype="bar", rwidth=0.8, color='Grey')
plt.xlabel("Price")
plt.ylabel("Mileage")
plt.title("German_Auto")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
slices = [50, 14, 24, 16]
Specification = ["Price", "Mileage", "Sales", "HorsePower"]
color = ["Green", "Grey", "Black", "Gold"]

plt.figure(figsize=(8, 6)) 
plt.pie(slices, labels=Specification, colors=color,
        startangle=90, explode=(0, 0.2, 0, 0),
        shadow=True, autopct="%.1f%%")

plt.title("German Auto Specifications")
plt.legend(Specification, title="Specifications", loc="upper right")

plt.show()
```

```python
import matplotlib.pyplot as plt
years = ["2015", "2016", "2017", "2018", "2019", "2020", "2021"]
profits = [9, 1, 10, 8, 2, 5, 6]
plt.figure(figsize=(10, 6))  
plt.plot(years, profits, marker='o', color='blue', linestyle='-', linewidth=2)
plt.title("Company Profits Year-wise", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Profit (in Rs)", fontsize=14)
plt.tight_layout()
plt.show()
```
